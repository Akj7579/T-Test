### Independent Sample - Unpaired T Test

Different Groups under same condition(salary)

```python
from scipy.stats import ttest_ind
df=df.dropna()
male = df[df['gender']=='M']['salary']
female = df[df['gender']=='F']['salary']
ttest_ind(male,female)
```

### Dependent Sample - Paired T Test

Same Group under different conditions

```python
from scipy.stats import ttest_rel
df=df.dropna()
male = df[df['gender']=='M']['ssc_p']
male1 = df[df['gender']=='M']['hsc_p']
ttest_rel(male,male1)
```

### CHECK THE RELATIONSHIP BETWEEN TENTH EXAM BOARDS AND SALARY

```python
from scipy.stats import ttest_ind  #import t test independent library
df=df.dropna()                     #dropping all null values
central_board = df[df['ssc_b']=='Central']['salary'] #query for getting salary for central board tenth exam
other_board = df[df['ssc_b']=='Others']['salary'] #query for getting salary for central board tenth exam
ttest_ind(central_board,other_board) #t test is done btw both
```

Since the p-value is greater than 0.05, it indicates that there is no significant difference between the two groups. This means that whether someone studied in a Central board or another board does not appear to have any meaningful impact on their salary.

### CHECK THE RELATIONSHIP BETWEEN HSC EXAM BOARDS AND ENTRANCE TEST MARK

```python
from scipy.stats import ttest_ind  #import t test independent library
df=df.dropna()                     #dropping all null values
central_board = df[df['hsc_b']=='Central']['etest_p'] #query for getting salary for central board tenth exam
other_board = df[df['hsc_b']=='Others']['etest_p'] #query for getting salary for central board tenth exam
ttest_ind(central_board,other_board) #t test is done btw both
```

### CHECK THE RELATIONSHIP BETWEEN HSC COMMERCE STUDENT'S MARK IN MBA EXAM AND ENTRANCE TEST

```python
from scipy.stats import ttest_rel #import t test dependent library
df=df.dropna()                    #dropping all null values
mba_passMark = df[df['hsc_s']=='Commerce']['mba_p']  #query for getting mba mark for hsc commerce student
entrance_passMark = df[df['hsc_s']=='Commerce']['etest_p']  #query for entarnce mark for hsc commerce student
ttest_rel(mba_passMark,entrance_passMark) #t test is done btw both
```

Since the p-value is far less than 0.05, it indicates a very strong and statistically significant difference between the two sets of marks. We can conclude that Commerce students’ performance in the MBA exam is significantly different from their performance in the entrance test.ie Their scores in one exam are not closely matching the other

### CHECK THE RELATIONSHIP BETWEEN SSC CENTRAL STUDENT'S MARK IN HSC AND DEGREE

```python
from scipy.stats import ttest_rel #import t test dependent library
df=df.dropna()                    #dropping all null values
hsc_mark = df[df['ssc_b']=='Central']['hsc_p']  #query for getting hsc percentage for central board tenth exam
degree_mark = df[df['ssc_b']=='Central']['degree_p']  #query for getting degree percenatge for central board tenth exam
ttest_rel(hsc_mark,degree_mark) #t test is done btw both
```

Since P Value is >0.05 and <0.10 it indicates a small difference between the two related samples, with the first group slightly higher, but it’s not strong enough to be statistically significant at the 5% level. ie There might be a slight improvement or decline, but we can’t confidently say that the marks in HSC and Degree are truly different — the change could simply be due to natural variation or other factors like Change in Curriculum Difficulty,Study Habits & Time Management etc
